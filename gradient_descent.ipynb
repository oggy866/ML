```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(starting_x,learning_rate,no_of_iterations):
    x = starting_x
    x_history = [x]
    y_history = [(x+3)**2]

    for i in range(no_of_iterations):
        gradient = 2*(x+3)
        x = x-learning_rate*gradient # code of gradient descent
        x_history.append(x)
        y_history.append((x+3)**2)

    return x_history,y_history
```

```python
starting_x = 2
learning_rate = 0.2
no_of_iterations = 10

x_history,y_history = gradient_descent(starting_x,learning_rate,no_of_iterations)
```

```python
x_vals = np.linspace(-15,10,400)
y_vals = (x_vals+3)**2

plt.plot(x_vals,y_vals)
```

```python
plt.scatter(x_history,y_history,label='Gradient Descent of (x+3)**2',color = 'red',marker='x')
plt.xlabel('x')
plt.ylabel('y')
# plt.legend()
plt.title('Gradient Descent of (x+3)**2')
plt.grid(True)
plt.show()
```
